```python
# ==============================================================================
# END-TO-END SALES FORECASTING & DEMAND INTELLIGENCE SYSTEM
# Internship Project (Week 3 & Week 4)
# Student: Prashant Ahuja
# ==============================================================================
# TASK 1: Data Loading, Merging & Deep Exploration
# ==============================================================================

# ==============================================================================
# Install Required Libraries
# ==============================================================================

!pip -q install prophet xgboost statsmodels plotly openpyxl python-docx joblib

# ==============================================================================
# Import Required Libraries
# ==============================================================================

import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go

from datetime import datetime

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

# ==============================================================================
# Mount Google Drive
# ==============================================================================

from google.colab import drive
drive.mount('/content/drive')

# ==============================================================================
# Project Directory
# ==============================================================================

PROJECT_PATH = "/content/drive/MyDrive/SalesForecasting_PrashantAhuja"

TRAIN_PATH = os.path.join(PROJECT_PATH, "train.csv")

CHART_PATH = os.path.join(PROJECT_PATH, "charts")
OUTPUT_PATH = os.path.join(PROJECT_PATH, "outputs")
MODEL_PATH = os.path.join(PROJECT_PATH, "models")

os.makedirs(CHART_PATH, exist_ok=True)
os.makedirs(OUTPUT_PATH, exist_ok=True)
os.makedirs(MODEL_PATH, exist_ok=True)

print("Project Folder :", PROJECT_PATH)

# ==============================================================================
# Load Superstore Sales Dataset
# ==============================================================================

sales = pd.read_csv(TRAIN_PATH)

sales["Order Date"] = pd.to_datetime(
    sales["Order Date"],
    format="%d/%m/%Y"
)

sales["Ship Date"] = pd.to_datetime(
    sales["Ship Date"],
    format="%d/%m/%Y"
)

print("\nDataset Loaded Successfully")
print("-" * 60)
print(f"Rows    : {sales.shape[0]}")
print(f"Columns : {sales.shape[1]}")

display(sales.head())

# ==============================================================================
# Dataset Information
# ==============================================================================

print("\nDataset Information")
print("-" * 60)

sales.info()

# ==============================================================================
# Missing Values
# ==============================================================================

missing_values = pd.DataFrame({
    "Missing Values": sales.isnull().sum(),
    "Percentage": round((sales.isnull().sum()/len(sales))*100,2)
})

display(missing_values)

# ==============================================================================
# Duplicate Records
# ==============================================================================

duplicate_rows = sales.duplicated().sum()

print(f"Duplicate Rows : {duplicate_rows}")

if duplicate_rows > 0:
    sales.drop_duplicates(inplace=True)
    print("Duplicate rows removed successfully.")
else:
    print("No duplicate rows found.")

print("\nFinal Dataset Shape :", sales.shape)

# ==============================================================================
# Data Type Validation
# ==============================================================================

display(pd.DataFrame(sales.dtypes, columns=["Data Type"]))

# ==============================================================================
# Feature Engineering
# ==============================================================================

sales["Year"] = sales["Order Date"].dt.year
sales["Month"] = sales["Order Date"].dt.month
sales["Month Name"] = sales["Order Date"].dt.month_name()
sales["Week Number"] = sales["Order Date"].dt.isocalendar().week.astype(int)
sales["Day"] = sales["Order Date"].dt.day
sales["Day of Week"] = sales["Order Date"].dt.day_name()
sales["Quarter"] = sales["Order Date"].dt.quarter

def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

sales["Season"] = sales["Month"].apply(get_season)

sales["Shipping Days"] = (
    sales["Ship Date"] - sales["Order Date"]
).dt.days

print("\nNew Features Added Successfully")

display(sales.head())

# ==============================================================================
# Save Clean Dataset
# ==============================================================================

sales.to_csv(
    os.path.join(OUTPUT_PATH, "cleaned_sales.csv"),
    index=False
)

print("Clean dataset saved successfully.")

# ==============================================================================
# Daily Sales Aggregation
# ==============================================================================

daily_sales = (
    sales.groupby("Order Date")["Sales"]
    .sum()
    .reset_index()
    .rename(columns={"Sales":"Daily Sales"})
)

daily_sales.to_csv(
    os.path.join(OUTPUT_PATH, "daily_sales.csv"),
    index=False
)

# ==============================================================================
# Weekly Sales Aggregation
# ==============================================================================

weekly_sales = (
    sales.resample("W", on="Order Date")["Sales"]
    .sum()
    .reset_index()
    .rename(columns={"Sales":"Weekly Sales"})
)

weekly_sales.to_csv(
    os.path.join(OUTPUT_PATH, "weekly_sales.csv"),
    index=False
)

# ==============================================================================
# Monthly Sales Aggregation
# ==============================================================================

monthly_sales = (
    sales.resample("M", on="Order Date")["Sales"]
    .sum()
    .reset_index()
    .rename(columns={"Sales":"Monthly Sales"})
)

monthly_sales.to_csv(
    os.path.join(OUTPUT_PATH, "monthly_sales.csv"),
    index=False
)

print("\nDaily, Weekly and Monthly aggregations created successfully.")

display(monthly_sales.head())
# ==============================================================================
# Exploratory Data Analysis (EDA)
# ==============================================================================

# ------------------------------------------------------------------------------
# Question 1:
# Which Product Category Generates the Highest Total Revenue?
# ------------------------------------------------------------------------------

category_sales = (
    sales.groupby("Category")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

print("\nTotal Revenue by Category")
display(category_sales)

plt.figure(figsize=(8,5))

sns.barplot(
    data=category_sales,
    x="Category",
    y="Sales",
    palette="viridis"
)

plt.title("Total Revenue by Product Category", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Revenue")

for i, value in enumerate(category_sales["Sales"]):
    plt.text(
        i,
        value,
        f"{value:,.0f}",
        ha="center",
        fontsize=10
    )

plt.tight_layout()

plt.savefig(
    os.path.join(CHART_PATH, "category_revenue.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

print(f"\nHighest Revenue Category : {category_sales.iloc[0]['Category']}")

# ------------------------------------------------------------------------------
# Question 2:
# Which Region Shows the Most Consistent Sales Growth?
# ------------------------------------------------------------------------------

region_year_sales = sales.pivot_table(
    values="Sales",
    index="Year",
    columns="Region",
    aggfunc="sum"
)

display(region_year_sales)

plt.figure(figsize=(10,6))

for region in region_year_sales.columns:

    plt.plot(
        region_year_sales.index,
        region_year_sales[region],
        marker="o",
        linewidth=2,
        label=region
    )

plt.title("Year-wise Sales Growth by Region")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(alpha=0.3)
plt.legend()

plt.tight_layout()

plt.savefig(
    os.path.join(CHART_PATH, "region_growth.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

growth_variation = region_year_sales.pct_change().std()

consistent_region = growth_variation.idxmin()

print("\nGrowth Variation")
display(growth_variation)

print(f"\nMost Consistent Sales Growth Region : {consistent_region}")

# ------------------------------------------------------------------------------
# Question 3:
# Average Shipping Time by Region
# ------------------------------------------------------------------------------

shipping_analysis = (
    sales.groupby("Region")["Shipping Days"]
    .mean()
    .sort_values()
    .reset_index()
)

display(shipping_analysis)

plt.figure(figsize=(8,5))

sns.barplot(
    data=shipping_analysis,
    x="Region",
    y="Shipping Days",
    palette="coolwarm"
)

plt.title("Average Shipping Time by Region")
plt.xlabel("Region")
plt.ylabel("Average Shipping Days")

for i, value in enumerate(shipping_analysis["Shipping Days"]):
    plt.text(
        i,
        value,
        f"{value:.1f}",
        ha="center"
    )

plt.tight_layout()

plt.savefig(
    os.path.join(CHART_PATH, "shipping_days_region.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

overall_shipping = sales["Shipping Days"].mean()

print(f"\nOverall Average Shipping Time : {overall_shipping:.2f} Days")

# ------------------------------------------------------------------------------
# Question 4:
# Are There Months That Consistently Spike Across All Years?
# ------------------------------------------------------------------------------

month_order = [
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
]

seasonality = (
    sales.groupby(["Year","Month Name"])["Sales"]
    .sum()
    .reset_index()
)

seasonality["Month Name"] = pd.Categorical(
    seasonality["Month Name"],
    categories=month_order,
    ordered=True
)

seasonality = seasonality.sort_values(
    ["Year","Month Name"]
)

plt.figure(figsize=(15,6))

sns.lineplot(
    data=seasonality,
    x="Month Name",
    y="Sales",
    hue="Year",
    marker="o",
    linewidth=2
)

plt.title("Monthly Sales Trend Across All Years")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(
    os.path.join(CHART_PATH, "monthly_seasonality.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

monthly_avg = (
    seasonality.groupby("Month Name")["Sales"]
    .mean()
    .sort_values(ascending=False)
)

print("\nAverage Sales by Month")
display(monthly_avg)

print(f"\nHighest Average Sales Month : {monthly_avg.index[0]}")

# ------------------------------------------------------------------------------
# Correlation Matrix
# ------------------------------------------------------------------------------

numeric_columns = sales.select_dtypes(include=np.number)

plt.figure(figsize=(10,7))

sns.heatmap(
    numeric_columns.corr(),
    annot=True,
    cmap="RdYlBu",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.tight_layout()

plt.savefig(
    os.path.join(CHART_PATH, "correlation_matrix.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# ------------------------------------------------------------------------------
# Save EDA Results
# ------------------------------------------------------------------------------

category_sales.to_csv(
    os.path.join(OUTPUT_PATH, "category_sales.csv"),
    index=False
)

region_year_sales.to_csv(
    os.path.join(OUTPUT_PATH, "region_year_sales.csv")
)

shipping_analysis.to_csv(
    os.path.join(OUTPUT_PATH, "shipping_analysis.csv"),
    index=False
)

seasonality.to_csv(
    os.path.join(OUTPUT_PATH, "seasonality_analysis.csv"),
    index=False
)

print("=" * 70)
print("TASK 1 COMPLETED SUCCESSFULLY")
print("=" * 70)

print("\nGenerated Charts")
print("- category_revenue.png")
print("- region_growth.png")
print("- shipping_days_region.png")
print("- monthly_seasonality.png")
print("- correlation_matrix.png")

print("\nGenerated Output Files")
print("- cleaned_sales.csv")
print("- daily_sales.csv")
print("- weekly_sales.csv")
print("- monthly_sales.csv")
print("- category_sales.csv")
print("- region_year_sales.csv")
print("- shipping_analysis.csv")
print("- seasonality_analysis.csv")
```

```python
# ==============================================================================
# TASK 2 : TIME SERIES ANALYSIS & DECOMPOSITION
# ==============================================================================

print("="*80)
print("TASK 2 : TIME SERIES ANALYSIS & DECOMPOSITION")
print("="*80)

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# ------------------------------------------------------------------------------
# Monthly Sales Time Series
# ------------------------------------------------------------------------------

monthly_ts = (
    sales
    .groupby(pd.Grouper(key="Order Date", freq="M"))["Sales"]
    .sum()
    .reset_index()
)

monthly_ts.columns = ["Order Date","Sales"]

display(monthly_ts.head())

monthly_ts.to_csv(
    os.path.join(OUTPUT_PATH,"monthly_time_series.csv"),
    index=False
)

# ------------------------------------------------------------------------------
# Plot Monthly Sales Trend
# ------------------------------------------------------------------------------

plt.figure(figsize=(16,6))

plt.plot(
    monthly_ts["Order Date"],
    monthly_ts["Sales"],
    linewidth=2,
    color="royalblue"
)

plt.title("Overall Monthly Sales Trend (4 Years)",fontsize=16)

plt.xlabel("Order Date")

plt.ylabel("Monthly Sales")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(
    os.path.join(
        CHART_PATH,
        "monthly_sales_trend.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# ------------------------------------------------------------------------------
# Prepare Time Series
# ------------------------------------------------------------------------------

ts = monthly_ts.copy()

ts.set_index(
    "Order Date",
    inplace=True
)

display(ts.head())

# ------------------------------------------------------------------------------
# Time Series Decomposition
# ------------------------------------------------------------------------------

decomposition = seasonal_decompose(
    ts["Sales"],
    model="additive",
    period=12
)

trend = decomposition.trend

seasonal = decomposition.seasonal

residual = decomposition.resid

# ------------------------------------------------------------------------------
# Plot All Components
# ------------------------------------------------------------------------------

fig = decomposition.plot()

fig.set_size_inches(16,10)

plt.tight_layout()

plt.savefig(
    os.path.join(
        CHART_PATH,
        "time_series_decomposition.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# ------------------------------------------------------------------------------
# Save Components
# ------------------------------------------------------------------------------

components = pd.DataFrame({

    "Sales":ts["Sales"],

    "Trend":trend,

    "Seasonality":seasonal,

    "Residual":residual

})

components.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "decomposition_components.csv"
    )

)

display(components.head(15))

# ------------------------------------------------------------------------------
# Trend Analysis
# ------------------------------------------------------------------------------

print("="*80)

print("TREND ANALYSIS")

print("="*80)

overall_growth = (
    ts["Sales"].iloc[-1] -
    ts["Sales"].iloc[0]
)

if overall_growth > 0:

    print("Overall Trend : Increasing")

elif overall_growth < 0:

    print("Overall Trend : Decreasing")

else:

    print("Overall Trend : Stable")

print()

print("Average Monthly Sales")

print(round(ts["Sales"].mean(),2))

print()

print("Maximum Monthly Sales")

print(round(ts["Sales"].max(),2))

print()

print("Minimum Monthly Sales")

print(round(ts["Sales"].min(),2))

# ------------------------------------------------------------------------------
# Seasonality Analysis
# ------------------------------------------------------------------------------

print("\n")

print("="*80)

print("SEASONALITY ANALYSIS")

print("="*80)

seasonality_strength = seasonal.abs().mean()

print("Average Seasonal Effect")

print(round(seasonality_strength,2))

highest_month = (

    monthly_ts
    .assign(
        Month=monthly_ts["Order Date"].dt.month_name()
    )
    .groupby("Month")["Sales"]
    .mean()
    .sort_values(
        ascending=False
    )

)

print("\nAverage Sales by Month")

display(highest_month)

# ------------------------------------------------------------------------------
# Residual Analysis
# ------------------------------------------------------------------------------

print("\n")

print("="*80)

print("RESIDUAL ANALYSIS")

print("="*80)

largest_noise = residual.abs().sort_values(
    ascending=False
)

print("Top Residual Months")

display(largest_noise.head(10))

plt.figure(figsize=(14,5))

plt.plot(
    residual,
    color="darkorange"
)

plt.axhline(
    0,
    color="black",
    linestyle="--"
)

plt.title("Residual Component")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "residual_component.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ------------------------------------------------------------------------------
# Augmented Dickey-Fuller Test
# ------------------------------------------------------------------------------

print("\n")

print("="*80)

print("AUGMENTED DICKEY FULLER TEST")

print("="*80)

adf_result = adfuller(
    ts["Sales"]
)

labels = [

    "ADF Statistic",

    "p-value",

    "Lags Used",

    "Number of Observations"

]

for value,label in zip(adf_result[:4],labels):

    print(f"{label} : {value}")

print()

print("Critical Values")

for key,value in adf_result[4].items():

    print(key," : ",value)

print()

if adf_result[1] < 0.05:

    print("Conclusion :")

    print("The time series is stationary.")

else:

    print("Conclusion :")

    print("The time series is NOT stationary.")

print()

print("Plain English Interpretation:")

print("A stationary time series has constant statistical properties such as mean and variance over time.")

print("Forecasting models like SARIMA usually perform better on stationary data.")

# ------------------------------------------------------------------------------
# Apply First Order Differencing
# ------------------------------------------------------------------------------

ts_diff = ts.copy()

ts_diff["Sales"] = ts_diff["Sales"].diff()

ts_diff = ts_diff.dropna()

display(ts_diff.head())

ts_diff.to_csv(
    os.path.join(
        OUTPUT_PATH,
        "differenced_time_series.csv"
    )
)

# ------------------------------------------------------------------------------
# Plot Original vs Differenced Series
# ------------------------------------------------------------------------------

plt.figure(figsize=(16,6))

plt.plot(
    ts.index,
    ts["Sales"],
    label="Original Series",
    linewidth=2
)

plt.plot(
    ts_diff.index,
    ts_diff["Sales"],
    label="Differenced Series",
    linewidth=2
)

plt.title("Original vs Differenced Monthly Sales")

plt.xlabel("Order Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(
    os.path.join(
        CHART_PATH,
        "original_vs_differenced.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# ------------------------------------------------------------------------------
# ADF Test After Differencing
# ------------------------------------------------------------------------------

print("\n")
print("="*80)
print("ADF TEST AFTER DIFFERENCING")
print("="*80)

adf_diff = adfuller(ts_diff["Sales"])

labels = [
    "ADF Statistic",
    "p-value",
    "Lags Used",
    "Number of Observations"
]

for value, label in zip(adf_diff[:4], labels):
    print(f"{label} : {value}")

print("\nCritical Values")

for key, value in adf_diff[4].items():
    print(f"{key} : {value}")

if adf_diff[1] < 0.05:
    print("\nConclusion : The differenced series is stationary.")
else:
    print("\nConclusion : The differenced series is still NOT stationary.")

# ------------------------------------------------------------------------------
# Compare Before and After Differencing
# ------------------------------------------------------------------------------

comparison = pd.DataFrame({
    "Before Differencing": [
        adf_result[0],
        adf_result[1]
    ],
    "After Differencing": [
        adf_diff[0],
        adf_diff[1]
    ]
},
index=["ADF Statistic","p-value"])

display(comparison)

comparison.to_csv(
    os.path.join(
        OUTPUT_PATH,
        "adf_test_comparison.csv"
    )
)

# ------------------------------------------------------------------------------
# Final Observations
# ------------------------------------------------------------------------------

print("\n")
print("="*80)
print("OBSERVATIONS")
print("="*80)

print("\n1. Trend")
print("- Monthly sales show the long-term movement in business performance.")
print("- An increasing trend indicates improving sales over the years.")

print("\n2. Seasonality")
print("- Monthly patterns repeat every year, indicating seasonal demand.")
print("- High seasonal months should be considered during inventory planning.")

print("\n3. Residual Component")
print("- Residuals represent random fluctuations that are not explained")
print("  by trend or seasonality.")
print("- Large residual values may indicate promotions, holidays or")
print("  unexpected business events.")

print("\n4. Stationarity")
if adf_diff[1] < 0.05:
    print("- After differencing, the series became stationary.")
    print("- The transformed series is suitable for SARIMA modelling.")
else:
    print("- Additional transformations may be required before modelling.")

# ------------------------------------------------------------------------------
# Save Final Time Series Data
# ------------------------------------------------------------------------------

ts.to_csv(
    os.path.join(
        OUTPUT_PATH,
        "monthly_sales_series.csv"
    )
)

print("\nTime series datasets saved successfully.")

# ------------------------------------------------------------------------------
# Charts Generated
# ------------------------------------------------------------------------------

print("\nCharts Saved")

charts = [
    "monthly_sales_trend.png",
    "time_series_decomposition.png",
    "residual_component.png",
    "original_vs_differenced.png"
]

for chart in charts:
    print(f"✔ {chart}")

# ------------------------------------------------------------------------------
# Output Files Generated
# ------------------------------------------------------------------------------

print("\nOutput Files Saved")

outputs = [
    "monthly_time_series.csv",
    "decomposition_components.csv",
    "differenced_time_series.csv",
    "monthly_sales_series.csv",
    "adf_test_comparison.csv"
]

for file in outputs:
    print(f"✔ {file}")

print("\n" + "="*80)
print("TASK 2 COMPLETED SUCCESSFULLY")
print("="*80)
```

```python
# ==============================================================================
# TASK 3 : SALES FORECASTING USING 3 DIFFERENT MODELS
# MODEL 1 : SARIMA
# MODEL 2 : FACEBOOK PROPHET
# ==============================================================================

print("="*80)
print("TASK 3 : SALES FORECASTING")
print("="*80)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

from statsmodels.tsa.statespace.sarimax import SARIMAX

from prophet import Prophet

import joblib

# ------------------------------------------------------------------------------
# Prepare Monthly Time Series
# ------------------------------------------------------------------------------

forecast_df = (
    sales
    .groupby(pd.Grouper(key="Order Date",freq="M"))["Sales"]
    .sum()
    .reset_index()
)

forecast_df.columns = ["Date","Sales"]

display(forecast_df.head())

forecast_df.to_csv(
    os.path.join(
        OUTPUT_PATH,
        "forecast_dataset.csv"
    ),
    index=False
)

# ------------------------------------------------------------------------------
# Train Test Split
# ------------------------------------------------------------------------------

train_size = int(len(forecast_df)*0.80)

train = forecast_df.iloc[:train_size]

test = forecast_df.iloc[train_size:]

print("Training Samples :",len(train))
print("Testing Samples :",len(test))

# ------------------------------------------------------------------------------
# Evaluation Metric Function
# ------------------------------------------------------------------------------

def calculate_metrics(actual,predicted):

    mae = mean_absolute_error(actual,predicted)

    rmse = np.sqrt(
        mean_squared_error(actual,predicted)
    )

    mape = np.mean(
        np.abs(
            (actual-predicted)/actual
        )
    )*100

    return mae,rmse,mape

# ==============================================================================
# MODEL 1 : SARIMA
# ==============================================================================

print("\n")
print("="*80)
print("MODEL 1 : SARIMA")
print("="*80)

sarima_train = train.set_index("Date")

sarima_test = test.set_index("Date")

sarima_model = SARIMAX(

    sarima_train["Sales"],

    order=(1,1,1),

    seasonal_order=(1,1,1,12),

    enforce_stationarity=False,

    enforce_invertibility=False

)

sarima_fit = sarima_model.fit()

print(sarima_fit.summary())

joblib.dump(

    sarima_fit,

    os.path.join(
        MODEL_PATH,
        "sarima_model.pkl"
    )

)

# ------------------------------------------------------------------------------
# Test Prediction
# ------------------------------------------------------------------------------

sarima_prediction = sarima_fit.predict(

    start=sarima_test.index[0],

    end=sarima_test.index[-1]

)

sarima_mae,sarima_rmse,sarima_mape = calculate_metrics(

    sarima_test["Sales"],

    sarima_prediction

)

print("\nPerformance")

print(f"MAE  : {sarima_mae:.2f}")

print(f"RMSE : {sarima_rmse:.2f}")

print(f"MAPE : {sarima_mape:.2f}%")

# ------------------------------------------------------------------------------
# Future Forecast
# ------------------------------------------------------------------------------

future_sarima = sarima_fit.get_forecast(
    steps=3
)

sarima_forecast = future_sarima.predicted_mean

sarima_ci = future_sarima.conf_int()

sarima_output = pd.DataFrame({

    "Forecast":sarima_forecast,

    "Lower CI":sarima_ci.iloc[:,0],

    "Upper CI":sarima_ci.iloc[:,1]

})

display(sarima_output)

sarima_output.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "sarima_forecast.csv"
    )

)

# ------------------------------------------------------------------------------
# SARIMA Plot
# ------------------------------------------------------------------------------

plt.figure(figsize=(15,6))

plt.plot(

    sarima_train.index,

    sarima_train["Sales"],

    label="Train"

)

plt.plot(

    sarima_test.index,

    sarima_test["Sales"],

    label="Actual"

)

plt.plot(

    sarima_prediction.index,

    sarima_prediction,

    label="Predicted"

)

plt.plot(

    sarima_forecast.index,

    sarima_forecast,

    label="Future Forecast",

    linewidth=3

)

plt.fill_between(

    sarima_ci.index,

    sarima_ci.iloc[:,0],

    sarima_ci.iloc[:,1],

    alpha=0.2

)

plt.title("SARIMA Forecast")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "sarima_forecast.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# MODEL 2 : FACEBOOK PROPHET
# ==============================================================================

print("\n")
print("="*80)
print("MODEL 2 : FACEBOOK PROPHET")
print("="*80)

prophet_df = forecast_df.copy()

prophet_df.columns = [

    "ds",

    "y"

]

prophet_train = prophet_df.iloc[:train_size]

prophet_test = prophet_df.iloc[train_size:]

prophet_model = Prophet(

    yearly_seasonality=True,

    weekly_seasonality=False,

    daily_seasonality=False

)

prophet_model.fit(prophet_train)

joblib.dump(

    prophet_model,

    os.path.join(
        MODEL_PATH,
        "prophet_model.pkl"
    )

)

# ------------------------------------------------------------------------------
# Prediction
# ------------------------------------------------------------------------------

future_test = prophet_model.make_future_dataframe(

    periods=len(prophet_test),

    freq="M"

)

forecast_test = prophet_model.predict(

    future_test

)

prophet_prediction = forecast_test.tail(

    len(prophet_test)

)["yhat"].values

prophet_mae,prophet_rmse,prophet_mape = calculate_metrics(

    prophet_test["y"],

    prophet_prediction

)

print("\nPerformance")

print(f"MAE  : {prophet_mae:.2f}")

print(f"RMSE : {prophet_rmse:.2f}")

print(f"MAPE : {prophet_mape:.2f}%")

# ------------------------------------------------------------------------------
# 3 Month Forecast
# ------------------------------------------------------------------------------

future = prophet_model.make_future_dataframe(

    periods=3,

    freq="M"

)

future_forecast = prophet_model.predict(

    future

)

prophet_output = future_forecast.tail(3)[

    [

        "ds",

        "yhat",

        "yhat_lower",

        "yhat_upper"

    ]

]

display(prophet_output)

prophet_output.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "prophet_forecast.csv"
    ),

    index=False

)

# ------------------------------------------------------------------------------
# Forecast Plot
# ------------------------------------------------------------------------------

fig1 = prophet_model.plot(
    future_forecast
)

fig1.set_size_inches(15,6)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "prophet_forecast.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ------------------------------------------------------------------------------
# Trend & Seasonality Components
# ------------------------------------------------------------------------------

fig2 = prophet_model.plot_components(
    future_forecast
)

fig2.set_size_inches(15,8)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "prophet_components.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

print("\nWeekly Seasonality")

print("Disabled because data is aggregated monthly.")

print("\nYearly Seasonality")

print("Automatically learned by Prophet and shown above.")
# ==============================================================================
# MODEL 3 : XGBOOST FOR TIME SERIES FORECASTING
# ==============================================================================

print("\n")
print("="*80)
print("MODEL 3 : XGBOOST")
print("="*80)

from xgboost import XGBRegressor

# ------------------------------------------------------------------------------
# Feature Engineering
# ------------------------------------------------------------------------------

xgb_df = forecast_df.copy()

xgb_df["Lag_1"] = xgb_df["Sales"].shift(1)
xgb_df["Lag_2"] = xgb_df["Sales"].shift(2)
xgb_df["Lag_3"] = xgb_df["Sales"].shift(3)

xgb_df["Rolling_Mean_3"] = (
    xgb_df["Sales"]
    .rolling(3)
    .mean()
)

xgb_df["Month"] = xgb_df["Date"].dt.month
xgb_df["Quarter"] = xgb_df["Date"].dt.quarter

def season(month):

    if month in [12,1,2]:
        return 1
    elif month in [3,4,5]:
        return 2
    elif month in [6,7,8]:
        return 3
    else:
        return 4

xgb_df["Season"] = xgb_df["Month"].apply(season)

xgb_df.dropna(inplace=True)

display(xgb_df.head())

# ------------------------------------------------------------------------------
# Train Test Split
# ------------------------------------------------------------------------------

train_size = int(len(xgb_df)*0.80)

train_xgb = xgb_df.iloc[:train_size]

test_xgb = xgb_df.iloc[train_size:]

features = [
    "Lag_1",
    "Lag_2",
    "Lag_3",
    "Rolling_Mean_3",
    "Month",
    "Quarter",
    "Season"
]

X_train = train_xgb[features]
y_train = train_xgb["Sales"]

X_test = test_xgb[features]
y_test = test_xgb["Sales"]

# ------------------------------------------------------------------------------
# Train Model
# ------------------------------------------------------------------------------

xgb_model = XGBRegressor(

    n_estimators=300,

    learning_rate=0.05,

    max_depth=4,

    random_state=42

)

xgb_model.fit(
    X_train,
    y_train
)

joblib.dump(

    xgb_model,

    os.path.join(
        MODEL_PATH,
        "xgboost_model.pkl"
    )

)

# ------------------------------------------------------------------------------
# Prediction
# ------------------------------------------------------------------------------

xgb_prediction = xgb_model.predict(X_test)

xgb_mae,xgb_rmse,xgb_mape = calculate_metrics(

    y_test,

    xgb_prediction

)

print("\nPerformance")

print(f"MAE  : {xgb_mae:.2f}")
print(f"RMSE : {xgb_rmse:.2f}")
print(f"MAPE : {xgb_mape:.2f}%")

# ------------------------------------------------------------------------------
# Future Forecast (3 Months)
# ------------------------------------------------------------------------------

future_data = xgb_df.copy()

future_predictions = []

future_dates = []

for i in range(3):

    last = future_data.iloc[-1]

    next_date = last["Date"] + pd.DateOffset(months=1)

    lag1 = last["Sales"]

    lag2 = future_data.iloc[-2]["Sales"]

    lag3 = future_data.iloc[-3]["Sales"]

    rolling = np.mean([lag1,lag2,lag3])

    month = next_date.month

    quarter = next_date.quarter

    season_value = season(month)

    future_input = pd.DataFrame({

        "Lag_1":[lag1],

        "Lag_2":[lag2],

        "Lag_3":[lag3],

        "Rolling_Mean_3":[rolling],

        "Month":[month],

        "Quarter":[quarter],

        "Season":[season_value]

    })

    prediction = xgb_model.predict(future_input)[0]

    future_predictions.append(prediction)

    future_dates.append(next_date)

    new_row = pd.DataFrame({

        "Date":[next_date],

        "Sales":[prediction],

        "Lag_1":[lag1],

        "Lag_2":[lag2],

        "Lag_3":[lag3],

        "Rolling_Mean_3":[rolling],

        "Month":[month],

        "Quarter":[quarter],

        "Season":[season_value]

    })

    future_data = pd.concat(
        [future_data,new_row],
        ignore_index=True
    )

future_forecast_xgb = pd.DataFrame({

    "Date":future_dates,

    "Forecast":future_predictions

})

display(future_forecast_xgb)

future_forecast_xgb.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "xgboost_forecast.csv"
    ),

    index=False

)

# ------------------------------------------------------------------------------
# Plot Forecast
# ------------------------------------------------------------------------------

plt.figure(figsize=(15,6))

plt.plot(

    train_xgb["Date"],

    train_xgb["Sales"],

    label="Train"

)

plt.plot(

    test_xgb["Date"],

    y_test,

    label="Actual"

)

plt.plot(

    test_xgb["Date"],

    xgb_prediction,

    label="Predicted"

)

plt.plot(

    future_dates,

    future_predictions,

    marker="o",

    linewidth=3,

    label="Future Forecast"

)

plt.title("XGBoost Forecast")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "xgboost_forecast.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# MODEL COMPARISON
# ==============================================================================

comparison = pd.DataFrame({

    "Model":[
        "SARIMA",
        "Prophet",
        "XGBoost"
    ],

    "MAE":[
        sarima_mae,
        prophet_mae,
        xgb_mae
    ],

    "RMSE":[
        sarima_rmse,
        prophet_rmse,
        xgb_rmse
    ],

    "MAPE":[
        sarima_mape,
        prophet_mape,
        xgb_mape
    ],

    "Forecast Month 1":[
        sarima_forecast.iloc[0],
        prophet_output.iloc[0]["yhat"],
        future_predictions[0]
    ],

    "Forecast Month 2":[
        sarima_forecast.iloc[1],
        prophet_output.iloc[1]["yhat"],
        future_predictions[1]
    ],

    "Forecast Month 3":[
        sarima_forecast.iloc[2],
        prophet_output.iloc[2]["yhat"],
        future_predictions[2]
    ]

})

display(comparison)

comparison.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "model_comparison.csv"
    ),

    index=False

)

# ------------------------------------------------------------------------------
# Best Model Selection
# ------------------------------------------------------------------------------

best_model = comparison.sort_values(
    "RMSE"
).iloc[0]

print("\n")
print("="*80)
print("BEST MODEL")
print("="*80)

print(f"Recommended Model : {best_model['Model']}")

print(f"RMSE : {best_model['RMSE']:.2f}")

print(f"MAE  : {best_model['MAE']:.2f}")

print(f"MAPE : {best_model['MAPE']:.2f}%")

print("\nRecommendation")

print(
    f"{best_model['Model']} is recommended for production because "
    "it achieved the lowest forecasting error on the evaluation dataset."
)

# ------------------------------------------------------------------------------
# Generated Files
# ------------------------------------------------------------------------------

print("\nSaved Models")

print("- sarima_model.pkl")
print("- prophet_model.pkl")
print("- xgboost_model.pkl")

print("\nSaved Charts")

print("- sarima_forecast.png")
print("- prophet_forecast.png")
print("- prophet_components.png")
print("- xgboost_forecast.png")

print("\nSaved Outputs")

print("- sarima_forecast.csv")
print("- prophet_forecast.csv")
print("- xgboost_forecast.csv")
print("- model_comparison.csv")

print("\n" + "="*80)
print("TASK 3 COMPLETED SUCCESSFULLY")
print("="*80)
```

```python
# ==============================================================================
# TASK 4 : PRODUCT CATEGORY & REGION LEVEL FORECASTING
# ==============================================================================

print("="*80)
print("TASK 4 : PRODUCT CATEGORY & REGION LEVEL FORECASTING")
print("="*80)

# ------------------------------------------------------------------------------
# Identify Best Model from Task 3
# ------------------------------------------------------------------------------

selected_model = best_model["Model"]

print(f"\nBest Performing Model Selected : {selected_model}")

# ------------------------------------------------------------------------------
# Function to Forecast using Best Model
# ------------------------------------------------------------------------------

def forecast_segment(data, segment_name):

    monthly = (
        data
        .groupby(pd.Grouper(key="Order Date", freq="M"))["Sales"]
        .sum()
        .reset_index()
    )

    monthly.columns = ["Date", "Sales"]

    if len(monthly) < 24:

        print(f"{segment_name} : Not enough data for forecasting.")
        return None

    # ==========================================================
    # SARIMA
    # ==========================================================

    if selected_model == "SARIMA":

        model = SARIMAX(

            monthly.set_index("Date")["Sales"],

            order=(1,1,1),

            seasonal_order=(1,1,1,12),

            enforce_stationarity=False,

            enforce_invertibility=False

        )

        fitted = model.fit(disp=False)

        forecast = fitted.get_forecast(steps=3)

        prediction = forecast.predicted_mean

        return pd.DataFrame({

            "Date":prediction.index,

            "Forecast":prediction.values,

            "Segment":segment_name

        })

    # ==========================================================
    # Prophet
    # ==========================================================

    elif selected_model == "Prophet":

        prophet_data = monthly.rename(
            columns={
                "Date":"ds",
                "Sales":"y"
            }
        )

        model = Prophet(

            yearly_seasonality=True,

            weekly_seasonality=False,

            daily_seasonality=False

        )

        model.fit(prophet_data)

        future = model.make_future_dataframe(

            periods=3,

            freq="M"

        )

        forecast = model.predict(future)

        result = forecast.tail(3)[["ds","yhat"]]

        result.columns = ["Date","Forecast"]

        result["Segment"] = segment_name

        return result

    # ==========================================================
    # XGBoost
    # ==========================================================

    else:

        temp = monthly.copy()

        temp["Lag_1"] = temp["Sales"].shift(1)

        temp["Lag_2"] = temp["Sales"].shift(2)

        temp["Lag_3"] = temp["Sales"].shift(3)

        temp["Rolling_Mean_3"] = temp["Sales"].rolling(3).mean()

        temp["Month"] = temp["Date"].dt.month

        temp["Quarter"] = temp["Date"].dt.quarter

        temp["Season"] = temp["Month"].apply(season)

        temp.dropna(inplace=True)

        X = temp[[
            "Lag_1",
            "Lag_2",
            "Lag_3",
            "Rolling_Mean_3",
            "Month",
            "Quarter",
            "Season"
        ]]

        y = temp["Sales"]

        model = XGBRegressor(

            n_estimators=300,

            learning_rate=0.05,

            max_depth=4,

            random_state=42

        )

        model.fit(X,y)

        working = temp.copy()

        predictions = []

        dates = []

        for i in range(3):

            last = working.iloc[-1]

            next_date = last["Date"] + pd.DateOffset(months=1)

            lag1 = last["Sales"]

            lag2 = working.iloc[-2]["Sales"]

            lag3 = working.iloc[-3]["Sales"]

            rolling = np.mean([lag1,lag2,lag3])

            feature = pd.DataFrame({

                "Lag_1":[lag1],

                "Lag_2":[lag2],

                "Lag_3":[lag3],

                "Rolling_Mean_3":[rolling],

                "Month":[next_date.month],

                "Quarter":[next_date.quarter],

                "Season":[season(next_date.month)]

            })

            pred = model.predict(feature)[0]

            predictions.append(pred)

            dates.append(next_date)

            new_row = pd.DataFrame({

                "Date":[next_date],

                "Sales":[pred],

                "Lag_1":[lag1],

                "Lag_2":[lag2],

                "Lag_3":[lag3],

                "Rolling_Mean_3":[rolling],

                "Month":[next_date.month],

                "Quarter":[next_date.quarter],

                "Season":[season(next_date.month)]

            })

            working = pd.concat(
                [working,new_row],
                ignore_index=True
            )

        return pd.DataFrame({

            "Date":dates,

            "Forecast":predictions,

            "Segment":segment_name

        })

# ------------------------------------------------------------------------------
# Forecast Categories
# ------------------------------------------------------------------------------

furniture = forecast_segment(

    sales[sales["Category"]=="Furniture"],

    "Furniture"

)

technology = forecast_segment(

    sales[sales["Category"]=="Technology"],

    "Technology"

)

office = forecast_segment(

    sales[sales["Category"]=="Office Supplies"],

    "Office Supplies"

)

# ------------------------------------------------------------------------------
# Forecast Regions
# ------------------------------------------------------------------------------

west = forecast_segment(

    sales[sales["Region"]=="West"],

    "West"

)

east = forecast_segment(

    sales[sales["Region"]=="East"],

    "East"

)

# ------------------------------------------------------------------------------
# Combine Forecasts
# ------------------------------------------------------------------------------

forecast_list = [

    furniture,

    technology,

    office,

    west,

    east

]

forecast_results = pd.concat(
    forecast_list,
    ignore_index=True
)

display(forecast_results)

forecast_results.to_csv(

    os.path.join(
        OUTPUT_PATH,
        "segment_forecasts.csv"
    ),

    index=False

)

# ------------------------------------------------------------------------------
# Comparison Plot
# ------------------------------------------------------------------------------

plt.figure(figsize=(15,6))

for segment in forecast_results["Segment"].unique():

    temp = forecast_results[
        forecast_results["Segment"]==segment
    ]

    plt.plot(

        temp["Date"],

        temp["Forecast"],

        marker="o",

        linewidth=2,

        label=segment

    )

plt.title("3-Month Forecast Comparison")

plt.xlabel("Forecast Month")

plt.ylabel("Forecast Sales")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(
        CHART_PATH,
        "category_region_forecast.png"
    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ------------------------------------------------------------------------------
# Strongest Growth
# ------------------------------------------------------------------------------

growth_summary = (

    forecast_results

    .groupby("Segment")["Forecast"]

    .mean()

    .sort_values(ascending=False)

)

display(growth_summary)

print("\n")

print("="*80)

print("BUSINESS INSIGHT")

print("="*80)

print(f"\nStrongest Upcoming Growth : {growth_summary.index[0]}")

print(f"Average Forecast Sales : {growth_summary.iloc[0]:.2f}")

print("\nRecommendation")

print(
    f"Based on the forecasting results, the {growth_summary.index[0]} "
    "segment is expected to show the strongest growth over the next "
    "three months. Inventory planning and stock allocation should be "
    "prioritized for this segment."
)

print("\n")

print("Generated Files")

print("- segment_forecasts.csv")

print("- category_region_forecast.png")

print("\n" + "="*80)

print("TASK 4 COMPLETED SUCCESSFULLY")

print("="*80)
```

```python
# ==============================================================================
# TASK 5 : ANOMALY DETECTION IN SALES DATA
# ==============================================================================
# Isolation Forest + Z-Score
# Supplementary Dataset : Video Game Sales (Multi-source Analysis)
# ==============================================================================

print("="*80)
print("TASK 5 : ANOMALY DETECTION")
print("="*80)

from sklearn.ensemble import IsolationForest
from scipy.stats import zscore

VGSALES_PATH = os.path.join(PROJECT_PATH, "vgsales.csv")

# ------------------------------------------------------------------------------
# Load Supplementary Dataset
# ------------------------------------------------------------------------------

vgsales = pd.read_csv(VGSALES_PATH)

print("\nVideo Game Sales Dataset")

print("Rows :", vgsales.shape[0])

print("Columns :", vgsales.shape[1])

display(vgsales.head())

# ------------------------------------------------------------------------------
# Missing Values
# ------------------------------------------------------------------------------

print("\nMissing Values")

display(vgsales.isnull().sum())

vgsales.dropna(inplace=True)

print("\nShape After Removing Missing Values")

print(vgsales.shape)

# ------------------------------------------------------------------------------
# Prepare Supplementary Dataset
# ------------------------------------------------------------------------------

vgsales["Year"] = vgsales["Year"].astype(int)

vg_year = (

    vgsales

    .groupby("Year")["Global_Sales"]

    .sum()

    .reset_index()

)

vg_year.columns = [

    "Year",

    "VideoGame_Global_Sales"

]

display(vg_year.head())

# ------------------------------------------------------------------------------
# Prepare Superstore Weekly Sales
# ------------------------------------------------------------------------------

weekly_sales = (

    sales

    .groupby(

        pd.Grouper(

            key="Order Date",

            freq="W"

        )

    )["Sales"]

    .sum()

    .reset_index()

)

weekly_sales.columns = [

    "Week",

    "Weekly Sales"

]

weekly_sales["Year"] = weekly_sales["Week"].dt.year

display(weekly_sales.head())

# ------------------------------------------------------------------------------
# Multi-source Merge
# ------------------------------------------------------------------------------

merged_data = pd.merge(

    weekly_sales,

    vg_year,

    on="Year",

    how="left"

)

display(merged_data.head())

merged_data.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "merged_sales_vgsales.csv"

    ),

    index=False

)

print("\nMerged Dataset Created Successfully")

# ==============================================================================
# Isolation Forest
# ==============================================================================

print("\n")

print("="*80)

print("ISOLATION FOREST")

print("="*80)

iso = IsolationForest(

    contamination=0.05,

    random_state=42

)

merged_data["Isolation_Label"] = iso.fit_predict(

    merged_data[["Weekly Sales"]]

)

merged_data["Isolation_Anomaly"] = merged_data["Isolation_Label"].map(

    {

        1:"Normal",

        -1:"Anomaly"

    }

)

anomaly_iso = merged_data[

    merged_data["Isolation_Label"]==-1

]

print("\nTotal Isolation Forest Anomalies")

print(len(anomaly_iso))

display(anomaly_iso)

# ------------------------------------------------------------------------------
# Plot Isolation Forest
# ------------------------------------------------------------------------------

plt.figure(figsize=(16,6))

plt.plot(

    merged_data["Week"],

    merged_data["Weekly Sales"],

    linewidth=2,

    label="Weekly Sales"

)

plt.scatter(

    anomaly_iso["Week"],

    anomaly_iso["Weekly Sales"],

    color="red",

    s=80,

    label="Isolation Forest"

)

plt.title("Isolation Forest Anomaly Detection")

plt.xlabel("Week")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(

        CHART_PATH,

        "isolation_forest_anomalies.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# Z-Score Method
# ==============================================================================

print("\n")

print("="*80)

print("Z-SCORE METHOD")

print("="*80)

merged_data["Rolling Mean"] = (

    merged_data["Weekly Sales"]

    .rolling(4)

    .mean()

)

merged_data["Z Score"] = zscore(

    merged_data["Weekly Sales"]

)

merged_data["Z_Anomaly"] = np.where(

    abs(merged_data["Z Score"])>2,

    "Anomaly",

    "Normal"

)

zscore_anomaly = merged_data[

    merged_data["Z_Anomaly"]=="Anomaly"

]

print("\nTotal Z-Score Anomalies")

print(len(zscore_anomaly))

display(zscore_anomaly)

# ------------------------------------------------------------------------------
# Plot Z-Score
# ------------------------------------------------------------------------------

plt.figure(figsize=(16,6))

plt.plot(

    merged_data["Week"],

    merged_data["Weekly Sales"],

    linewidth=2,

    label="Weekly Sales"

)

plt.scatter(

    zscore_anomaly["Week"],

    zscore_anomaly["Weekly Sales"],

    color="orange",

    s=80,

    label="Z-Score"

)

plt.title("Z-Score Based Anomaly Detection")

plt.xlabel("Week")

plt.ylabel("Weekly Sales")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(

        CHART_PATH,

        "zscore_anomalies.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# Compare Both Methods
# ==============================================================================

comparison = merged_data[[

    "Week",

    "Weekly Sales",

    "Isolation_Anomaly",

    "Z_Anomaly"

]]

comparison.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "anomaly_comparison.csv"

    ),

    index=False

)

display(comparison.head(20))

agreement = (

    comparison["Isolation_Anomaly"]

    ==

    comparison["Z_Anomaly"]

).sum()

print("\n")

print("="*80)

print("COMPARISON")

print("="*80)

print(f"Total Weekly Records : {len(comparison)}")

print(f"Matching Results : {agreement}")

print(f"Different Results : {len(comparison)-agreement}")

print("\nInterpretation")

print("- Isolation Forest detects anomalies using machine learning.")

print("- Z-Score detects anomalies statistically based on standard deviation.")

print("- If both methods detect the same weeks, confidence in the anomaly increases.")

print("- Different results indicate that each method captures different anomaly patterns.")

# ==============================================================================
# Possible Business Explanations
# ==============================================================================

print("\n")

print("="*80)

print("POSSIBLE BUSINESS EXPLANATIONS")

print("="*80)

for _, row in anomaly_iso.head(5).iterrows():

    month = row["Week"].month

    if month in [11,12]:

        reason = "Likely festive season or holiday sale."

    elif month in [6,7]:

        reason = "Possible seasonal promotion."

    else:

        reason = "Unexpected demand fluctuation or inventory issue."

    print(

        f"{row['Week'].date()} : {reason}"

    )

# ==============================================================================
# Save Reports
# ==============================================================================

anomaly_iso.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "isolation_forest_report.csv"

    ),

    index=False

)

zscore_anomaly.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "zscore_report.csv"

    ),

    index=False

)

print("\nGenerated Charts")

print("- isolation_forest_anomalies.png")

print("- zscore_anomalies.png")

print("\nGenerated Output Files")

print("- merged_sales_vgsales.csv")

print("- anomaly_comparison.csv")

print("- isolation_forest_report.csv")

print("- zscore_report.csv")

print("\n" + "="*80)

print("TASK 5 COMPLETED SUCCESSFULLY")

print("="*80)
```

```python
# ==============================================================================
# TASK 6 : PRODUCT DEMAND SEGMENTATION USING CLUSTERING
# ==============================================================================

print("="*80)
print("TASK 6 : PRODUCT DEMAND SEGMENTATION")
print("="*80)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# ------------------------------------------------------------------------------
# Prepare Product Level Dataset
# ------------------------------------------------------------------------------

product_monthly = (

    sales

    .groupby(

        [

            "Sub-Category",

            "Year",

            "Month"

        ]

    )["Sales"]

    .sum()

    .reset_index()

)

# ------------------------------------------------------------------------------
# Total Sales Volume
# ------------------------------------------------------------------------------

total_sales = (

    product_monthly

    .groupby("Sub-Category")["Sales"]

    .sum()

)

# ------------------------------------------------------------------------------
# Sales Volatility
# ------------------------------------------------------------------------------

volatility = (

    product_monthly

    .groupby("Sub-Category")["Sales"]

    .std()

)

# ------------------------------------------------------------------------------
# Average Order Value
# ------------------------------------------------------------------------------

average_order = (

    sales

    .groupby("Sub-Category")["Sales"]

    .mean()

)

# ------------------------------------------------------------------------------
# Year-over-Year Growth
# ------------------------------------------------------------------------------

yearly_sales = (

    sales

    .groupby(

        [

            "Sub-Category",

            "Year"

        ]

    )["Sales"]

    .sum()

    .reset_index()

)

growth = yearly_sales.pivot(

    index="Sub-Category",

    columns="Year",

    values="Sales"

)

growth = growth.fillna(0)

growth["Growth Rate"] = (

    growth.iloc[:,-1] -

    growth.iloc[:,0]

) / growth.iloc[:,0].replace(0,np.nan)

growth["Growth Rate"] = growth["Growth Rate"].fillna(0)

# ------------------------------------------------------------------------------
# Final Clustering Dataset
# ------------------------------------------------------------------------------

cluster_df = pd.DataFrame({

    "Total Sales":total_sales,

    "Growth Rate":growth["Growth Rate"],

    "Volatility":volatility,

    "Average Order Value":average_order

})

cluster_df = cluster_df.fillna(0)

display(cluster_df)

cluster_df.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "clustering_dataset.csv"

    )

)

# ==============================================================================
# Feature Scaling
# ==============================================================================

scaler = StandardScaler()

scaled_features = scaler.fit_transform(cluster_df)

# ==============================================================================
# Elbow Method
# ==============================================================================

wcss = []

for k in range(1,11):

    model = KMeans(

        n_clusters=k,

        random_state=42,

        n_init=10

    )

    model.fit(scaled_features)

    wcss.append(model.inertia_)

plt.figure(figsize=(8,5))

plt.plot(

    range(1,11),

    wcss,

    marker="o"

)

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(

        CHART_PATH,

        "elbow_method.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# K-Means Clustering
# ==============================================================================

optimal_clusters = 4

kmeans = KMeans(

    n_clusters=optimal_clusters,

    random_state=42,

    n_init=10

)

cluster_df["Cluster"] = kmeans.fit_predict(

    scaled_features

)

# ==============================================================================
# Cluster Labels
# ==============================================================================

cluster_names = {

    0:"High Volume, Stable Demand",

    1:"Low Volume, High Volatility",

    2:"Growing Demand",

    3:"Declining Demand"

}

cluster_df["Demand Segment"] = (

    cluster_df["Cluster"]

    .map(cluster_names)

)

display(cluster_df)

# ==============================================================================
# PCA Visualization
# ==============================================================================

pca = PCA(

    n_components=2,

    random_state=42

)

pca_features = pca.fit_transform(

    scaled_features

)

cluster_df["PCA1"] = pca_features[:,0]

cluster_df["PCA2"] = pca_features[:,1]

plt.figure(figsize=(10,7))

sns.scatterplot(

    data=cluster_df,

    x="PCA1",

    y="PCA2",

    hue="Demand Segment",

    s=120

)

for index,row in cluster_df.iterrows():

    plt.text(

        row["PCA1"],

        row["PCA2"],

        index,

        fontsize=8

    )

plt.title("Demand Segmentation using PCA")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(

    os.path.join(

        CHART_PATH,

        "product_clusters.png"

    ),

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# ==============================================================================
# Cluster Summary
# ==============================================================================

summary = (

    cluster_df

    .groupby("Demand Segment")[

        [

            "Total Sales",

            "Growth Rate",

            "Volatility",

            "Average Order Value"

        ]

    ]

    .mean()

)

display(summary)

summary.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "cluster_summary.csv"

    )

)

cluster_df.to_csv(

    os.path.join(

        OUTPUT_PATH,

        "product_segments.csv"

    )

)

# ==============================================================================
# Stocking Strategy
# ==============================================================================

print("\n")

print("="*80)

print("RECOMMENDED STOCKING STRATEGY")

print("="*80)

strategies = {

    "High Volume, Stable Demand":
    "Maintain high inventory levels with regular replenishment.",

    "Low Volume, High Volatility":
    "Keep limited stock and monitor demand frequently.",

    "Growing Demand":
    "Gradually increase inventory to meet rising demand.",

    "Declining Demand":
    "Reduce stock levels and avoid over-ordering."

}

for segment in cluster_df["Demand Segment"].unique():

    print(f"\n{segment}")

    print(strategies[segment])

print("\nGenerated Charts")

print("- elbow_method.png")

print("- product_clusters.png")

print("\nGenerated Output Files")

print("- clustering_dataset.csv")

print("- cluster_summary.csv")

print("- product_segments.csv")

print("\n" + "="*80)

print("TASK 6 COMPLETED SUCCESSFULLY")

print("="*80)
```
